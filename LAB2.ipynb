1. Implemente una funci√≥n en Python que permita generar una funci√≥n escal√≥n unitario, definida en un intervalo ùëõ1‚â§ùëõ0‚â§ùëõ2

```python
import numpy as np
import matplotlib.pyplot as plt

def escalon(n0, n1, n2):

    # Genera x(n) = escalon(n-n0); n1 <= n <= n2

    # ----------------------------------------------

    n = np.arange(n1,n2+1) # Se crea el vector de muestras

    x = (n-n0) >= 0

    return [x,n]

# Ejemplo de uso:
output_x, output_n = escalon(2,-5,10)

# Gr√°fico
plt.figure(figsize=(10, 6))
plt.stem(output_n, output_x)
plt.title('Funci√≥n Escalon: x(n) = r(n-2)')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.grid(True)
plt.show()
```

2. Implemente una funci√≥n en Python que permita generar una funci√≥n rampa, definida en un intervalo ùëõ1‚â§ùëõ0‚â§ùëõ2 con una pendiente m.

```python
import numpy as np
import matplotlib.pyplot as plt

def Rampa(n0, n1, n2):
# Genera x(n) = R(n-n0); n1 <= n <= n2
# ----------------------------------------------
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = np.zeros(len(n))
 for i in range(len(n)):
  if n[i] < n0:
   x[i] = 0
  else:
   x[i] = n[i] - n0
 return [x,n]

output_x, output_n = Rampa(2,-5,10)

plt.figure(figsize=(10, 6))
plt.stem(output_n, output_x)
plt.title('Funci√≥n Rampa: x(n) = r(n-2)')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Intervalo n
n = np.arange(0,18)

# Secuencia ùë•[ùëõ]
x = np.zeros(len(n))
for i, val in enumerate(n):
  if 0 <= val <= 5:
    x[i] = val
  elif 6 <= val <= 11:
    x[i] = val - 5
  elif 12 <= val <= 17:
    x[i] = val - 10

#a. Derivada
derivada = np.diff(x)

#b. Gr√°fica
plt.figure(figsize=(10,8))

# Secuencia original
plt.subplot(2, 1, 1)
plt.stem(n, x)
plt.title('Secuencia original')
plt.grid(True)
plt.xlabel('n')
plt.ylabel('x(n)')
# Secuencia derivada
plt.subplot(2, 1, 2)
plt.stem(n[:-1], derivada)
plt.title('Derivada de la secuencia')
plt.grid(True)
plt.xlabel('n')
plt.ylabel('x\'(n)')

plt.tight_layout()
plt.show()
```

Se observa que en los intevalos de la rampa, la derivada es cte, adem√°s en los picos de la rampa, la derivada en ese punto tiende a menos infinito.
