1) Escribir una funci√≥n que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota m√≠nima, la m√°xima, media y la desviaci√≥n t√≠pica.

```python
def estadisticos_notas(diccionario_notas):
    notas = pd.Series(diccionario_notas)

    estadisticos = pd.Series({
        'Nota Minima': notas.min(),
        'Nota Maxima': notas.max(),
        'Media': notas.mean(),
        'Desviaci√≥n Tipica': notas.std()
    })
    return estadisticos

#Ejemplo
notas = {'Ana': 4.5, 'Juan': 4.0, 'Gabriela': 2.3, 'Sofia': 5.0, 'Elena': 1.5}
resultado = estadisticos_notas(notas)
print(resultado)
```

2) C√°lculo del √çndice de Masa Corporal (IMC) con Pandas.

```python
#a.
df = pd.read_csv('datos.csv', sep=';')

#b.
print("Primeras y ultimas filas")
print(df.head())
print(df.tail())

#c.
df = df.drop(columns=['Unnamed: 0'], errors='ignore')

#d - e
#NOTAS: LAS PALABRAS "Weight" y "Height" ESTABAN TRUCADAS DESDE EL DATAFRAME, POR LO QUE TOCA HACER LO MSIMO AQUI.
df['BMI'] = df['Height'] / ((df['Weight'] / 100) ** 2)

#f.
limites = [0, 18.5, 25.0, 30.0, float('inf')]
etiquetas = ['Bajo peso', 'Normal', 'Sobrepeso', 'Obesidad']
df['Categoria_IMC'] = pd.cut(df['BMI'], bins=limites, labels=etiquetas, right=False)
print("Primeras filas despues de calcular IMC y categorizar")
print(df.head())
```
